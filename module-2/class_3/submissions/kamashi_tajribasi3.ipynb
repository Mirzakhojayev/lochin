Calculating age and creating a new column

```python
current_year = 2026   #define the current year
df['birth_year'] = df['date_of_birth'].str[-4:].astype(int)    #Extract the birth year from 'date_of_birth' and convert them
df['age'] = current_year - df['birth_year'] #creating new age column
print("Updated Dataframe with Age column:")   #showing results
df[['full_name', 'date_of_birth', 'age']].head()  #showing first 5 lines of results
```

Logic with If-Else (Customer Classification)

```python
def classify_customer(age):  #1. Create a function to define the logic
    if age >= 60:
        return "Senior"
    elif age >= 30:
        return "Adult"
    else:
        return "Youth"

df['status'] = df['age'].apply(classify_customer)    #2. Apply this function to every row in the 'age' column
print("Final Dataframe with Status classification:") #3. Print the final results
df[['full_name', 'age', 'status']].head(10)
```

Value Counts to summarize the data and create a simple Bar Chart.

```python
import matplotlib.pyplot as plt   #bringing a powerful library as a 'the factory of libraries'
status_counts = df['status'].value_counts()   #1. Count how many people are in each category
print("Customer Statistics:") #2. Print the summary text
print(status_counts)
plt.figure(figsize=(8, 5))   #3. Create a Bar Chart to visualize the data
status_counts.plot(kind='bar', color=['skyblue', 'orange', 'lightgreen'])
plt.title('Distribution of Customer Categories', fontsize=14)  #4. Add titles and labels
plt.xlabel('Category', fontsize=12)
plt.ylabel('Number of People', fontsize=12)
plt.xticks(rotation=0) # Keeps the labels horizontal
plt.show()   #Show the plot
```

```python
# 1. Variables
current_year = 2026
senior_count = 0

# 2. Function: Decide who gets a special bonus
def get_bonus_message(age, name):
    if age >= 60:
        return f"Special 5% Interest Bonus for {name}!"
    else:
        return f"Standard Monthly Update for {name}."

# 3. Loop: Iterate through the data and apply logic
print("--- BANK NOTIFICATION SYSTEM ---")

for index, row in df.iterrows():
    # Calculate age using row variable
    person_age = current_year - int(row['date_of_birth'][-4:])
    person_name = row['full_name']

    # Use the function
    message = get_bonus_message(person_age, person_name)

    # Logic to count only Seniors
    if person_age >= 60:
        print(message)
        senior_count += 1

# 4. Final Summary
print("--------------------------------")
print(f"Total Bonus Notifications Sent: {senior_count}")
```

```python
# 1. Variables
current_year = 2026

# 2. Optimized Function with English logic and messages
def get_detailed_bonus(age, name):
    if age >= 60:
        return f"{name} (Age: {age}): 5% Senior Bonus."
    elif age >= 40: # If Python reaches here, age is definitely below 60
        return f"{name} (Age: {age}): 10% Entrepreneurship Bonus."
    elif age >= 20: # If Python reaches here, age is definitely below 40
        return f"{name} (Age: {age}): 15% Education and Vocational Training Bonus."
    else: # For all ages below 20 (1-19 years old)
        return f"{name} (Age: {age}): 20% School and Preschool Education Bonus."

# 3. Running the Loop
print("--- OPTIMIZED CUSTOMER BONUS LIST ---")

for index, row in df.iterrows():
    # Calculate age by slicing the last 4 characters of the birth date
    person_age = current_year - int(row['date_of_birth'][-4:])
    person_name = row['full_name']

    # Print the result using the English function
    print(get_detailed_bonus(person_age, person_name))

print("-------------------------------------")
print("Mission Complete!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. SETTINGS & VARIABLES
current_year = 2026
# Define the order of categories for consistent plotting
category_order = ["Senior (5%)", "Entrepreneur (10%)", "Education (15%)", "Junior/School (20%)"]

# 2. LOGIC FUNCTIONS
def get_bonus_message(age, name):
    """Returns a specific bonus message based on age."""
    if age >= 60:
        return f"{name} (Age: {age}): 5% Senior Bonus."
    elif age >= 40:
        return f"{name} (Age: {age}): 10% Entrepreneurship Bonus."
    elif age >= 20:
        return f"{name} (Age: {age}): 15% Education and Vocational Training Bonus."
    else:
        return f"{name} (Age: {age}): 20% School and Preschool Education Bonus."

def get_bonus_group(age):
    """Categorizes the age into a group name for visualization."""
    if age >= 60: return "Senior (5%)"
    elif age >= 40: return "Entrepreneur (10%)"
    elif age >= 20: return "Education (15%)"
    else: return "Junior/School (20%)"

# 3. PROCESSING DATA
# Creating the visual group column
df['bonus_group'] = df['age'].apply(get_bonus_group)

print("--- CUSTOMER BONUS NOTIFICATION LIST ---")
for index, row in df.iterrows():
    # Calling our English function
    print(get_detailed_bonus(row['age'], row['full_name']))

print("-" * 40)
print("Processing Completed Successfully!")

# 4. VISUALIZATION
plt.figure(figsize=(16, 8))
sns.set_style("whitegrid")

# --- CHART 1: BAR CHART ---
plt.subplot(1, 2, 1)
# Using hue and legend=False to fix the FutureWarning
sns.countplot(data=df, x='bonus_group', order=category_order,
              hue='bonus_group', palette='magma', legend=False)
plt.title('Customer Count by Bonus Category', fontsize=16, fontweight='bold')
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Customers', fontsize=12)
plt.xticks(rotation=15)

# --- CHART 2: PIE CHART ---
plt.subplot(1, 2, 2)
# Filtering data to avoid "Invalid vertices array" or empty array errors
counts = df['bonus_group'].value_counts().reindex(category_order).fillna(0)
counts_filtered = counts[counts > 0]

if not counts_filtered.empty:
    plt.pie(counts_filtered, labels=counts_filtered.index, autopct='%1.1f%%',
            startangle=140, colors=sns.color_palette('magma', n_colors=len(counts_filtered)),
            explode=[0.05] * len(counts_filtered), shadow=True)
    plt.title('Percentage Share of Bonus Groups', fontsize=16, fontweight='bold')
else:
    plt.text(0.5, 0.5, 'No Data Found for Charting', ha='center', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# 1. Ensure the 'bonus_group' column exists using our English logic
def get_bonus_group(age):
    if age >= 60: return "Senior (5%)"
    elif age >= 40: return "Entrepreneur (10%)"
    elif age >= 20: return "Education (15%)"
    else: return "Junior/School (20%)"

df['bonus_group'] = df['age'].apply(get_bonus_group)

# 2. Define the logical order for the table rows
category_order = ["Senior (5%)", "Entrepreneur (10%)", "Education (15%)", "Junior/School (20%)"]

# 3. Create the Summary Table with Count and Average Age
summary_table = df.groupby('bonus_group').agg(
    Customer_Count=('age', 'count'),
    Average_Age=('age', 'mean')
).reindex(category_order)

# 4. Clean up the table (Rounding the average age)
summary_table['Average_Age'] = summary_table['Average_Age'].round(1)

# 5. Display the result
print("--- STRATEGIC SEGMENTATION SUMMARY TABLE ---")
display(summary_table) # 'display' looks better than 'print' in Notebooks
```

```python
import pandas as pd

# 1. Define the static data for the business logic
logic_data = {
    'Category': ['Senior', 'Entrepreneur', 'Education', 'Junior/School'],
    'Age Range': ['60+', '40-59', '20-39', '1-19'],
    'Bonus %': ['5%', '10%', '15%', '20%'],
    'Target Goal': ['Retirement Support', 'Business Development', 'Vocational Training', 'Early Education']
}

# 2. Create a mapping function for the data grouping
def get_category_name(age):
    if age >= 60: return "Senior"
    elif age >= 40: return "Entrepreneur"
    elif age >= 20: return "Education"
    else: return "Junior/School"

# 3. Apply grouping and calculate counts from your existing 'df'
df['Category'] = df['age'].apply(get_category_name)
counts = df['Category'].value_counts()

# 4. Combine everything into the final Result Table
final_summary = pd.DataFrame(logic_data)
final_summary['Customer_Count'] = final_summary['Category'].map(counts).fillna(0).astype(int)

# 5. Print the Final Success Message
print("="*50)
print("     THE SELF-STUDY MISSION IS COMPLETED!     ")
print("="*50)
print("\nFINAL STRATEGIC SUMMARY TABLE:")
print(final_summary.to_string(index=False))
print("\n" + "="*50)
print("Status: System Optimized & Data Visualized.")
```
